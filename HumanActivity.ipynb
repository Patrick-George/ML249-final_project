```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Making an array out of the text file
with open("UCI HAR Dataset/features.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
feature_names = []
for i in content:
    feature_names.append(i.split()[1])
f.close()
```

```python
print(len(feature_names))
```

```python
#Making an array out of the text file

with open("UCI HAR Dataset/activity_labels.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
activity_labels = []
for i in content:
    activity_labels.append(i.split())
f.close()
```

```python
activity_labels
```

```python
np.array(activity_labels)[:,1]
```

```python
#Making an array out of the text file
with open("UCI HAR Dataset/train/X_train.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
x_train = []
#making the training set array
for i in content:
    x_train.append(i.split())
f.close()
```

```python
print(len(x_train))
print(len(x_train[0]))
```

```python
#Making an array out of the text file
with open("UCI HAR Dataset/train/y_train.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
y_train = []
for i in content:
    y_train.append(i.split())
f.close()
print(len(y_train))
```

```python
#Making an array out of the text file
with open("UCI HAR Dataset/train/subject_train.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
subject_train = content
f.close()
print(len(subject_train))
print(subject_train[:10])
```

```python
#Making an array out of the text file
with open("UCI HAR Dataset/test/X_test.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
x_test = []
for i in content:
    x_test.append(i.split())
f.close()
print(len(x_test),len(x_test[0]))


with open("UCI HAR Dataset/test/y_test.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
y_test = []
for i in content:
    y_test.append(i.split())
f.close()
print(len(y_test))
print(y_test[:10])

with open("UCI HAR Dataset/test/subject_test.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
subject_test = content
f.close()
print(len(subject_test))
print(subject_test[:10])
```

```python
#Making the array appropriate to load it into a pandas dataframe
a_test = np.hstack((y_test,x_test))
a_train = np.hstack((y_train,x_train))
print(a_train.shape,a_test.shape)
b = np.hstack((['activity_label'],feature_names))
b.shape
```

```python
#Making pandas dataframe
df_train = pd.DataFrame.from_records(a_train, columns = b)
df_test = pd.DataFrame.from_records(a_test, columns = b)
print(df_train.shape,df_test.shape)
print(len(feature_names))
```

```python
print(df_train.shape,df_test.shape)
# df_train
df_train[:10]
```

```python
#PCA for converting the dimensions into 2
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_train[feature_names])
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df_train[['activity_label']]], axis = 1)
finalDf.head(5)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = np.array(activity_labels)[:,0]
colors = ['r', 'g', 'b','c','m','y']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['activity_label'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 5)
ax.legend(targets)
ax.grid()
print(activity_labels)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier().fit(df_train[feature_names], df_train['activity_label'])

print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(df_test[feature_names], df_test['activity_label'])))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
rfc.fit(df_train[feature_names], df_train['activity_label'])

print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(rfc.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(rfc.score(df_test[feature_names], df_test['activity_label'])))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(df_train[feature_names], df_train['activity_label'])

print('Accuracy of Logistic regression classifier on training set: {:.6f}'
     .format(logreg.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of Logistic regression classifier on test set: {:.6f}'
     .format(logreg.score(df_test[feature_names], df_test['activity_label'])))
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(df_train[feature_names], df_train['activity_label'])
print('Accuracy of LDA classifier on training set: {:.6f}'
     .format(lda.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of LDA classifier on test set: {:.6f}'
     .format(lda.score(df_test[feature_names], df_test['activity_label'])))
```

```python
#SVC with linear kernel
from sklearn.svm import SVC
clf = SVC(gamma='auto',C = 1, kernel = 'linear')
clf.fit(df_train[feature_names], df_train['activity_label'])
print('Accuracy of SVC classifier on training set: {:.6f}'
     .format(clf.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of SVC classifier on test set: {:.6f}'
     .format(clf.score(df_test[feature_names], df_test['activity_label'])))
```

```python
#SVC with polynomial kernel
from sklearn.svm import SVC
clf = SVC(gamma='auto',C = 1, kernel = 'poly')
# clf.fit(X, y) 
clf.fit(df_train[feature_names], df_train['activity_label'])
print('Accuracy of SVC classifier on training set: {:.6f}'
     .format(clf.score(df_train[feature_names], df_train['activity_label'])))
print('Accuracy of SVC classifier on test set: {:.6f}'
     .format(clf.score(df_test[feature_names], df_test['activity_label'])))
```
